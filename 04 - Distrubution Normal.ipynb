## Big Data Distributions

### Question 1: Create an array with 100000 random numbers, and display them using a histogram with 100 bars!

```python
x = np.random.uniform(0.0, 5.0, 100000)
plt.hist(x, 100)
plt.show()
```

high itu batas atas dari interval output.<br>
size itu ukuran data ingin sampai berapa, seperti yang diatas ingin mengeluarkan array 100000 random number<br>

## Normal Data Distribution

```python
x = np.random.normal(5.0, 1.0, 100000)
plt.hist(x, 100)
plt.show()
```

insight: pada data diatas kita bisa mengetahui bahwa Modality yang terbentuk adalah Unimodal, dan Skewnessnya Symmetric

## Another Way To Code Normal Distribution in Python

```python
#mu = mean & sigma = std. deviation
mu, sigma = 0.5, 0.1
s = np.random.normal(0.5,0.1, 1000)

#create the bins and histogram
count, bins, ignored = plt. hist(s, 20, density = True)

# Plot the distribution curve
plt. plot(bins, 1/(sigma*np.sqrt(2*np.pi))*
          np.exp(-(bins-mu)**2/(2*sigma**2)), linewidth=3, color='y'
         )
plt.show()
```

insight: pada graph diatas, sudah bisa diketahui ada garis distribution curve yang Modalitynya berbentuk Unimodal dan Skewnessnya Symmetric
