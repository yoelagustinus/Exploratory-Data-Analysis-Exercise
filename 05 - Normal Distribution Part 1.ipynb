```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
```

## Normal Distribution Probability Density Function

```python
norm.pdf(x=1.0, loc=0, scale=1)
```

```python
fig, ax = plt.subplots()
x= np.arange(-4,4,0.001)
ax.plot(x, norm.pdf(x))
ax.set_ylim(0,0.45) # range
ax.axhline(y=0.24,xmax=0.61,color='r') # horizontal line
ax.axvline(x=1, ymax=0.53, color='r',alpha=0.5) # vertical line
xplot = ax.plot([1], [0.24], marker='o', markersize=15, color="red") # coordinate point
ax.set_yticks([]) # remove y axis label
ax.set_xticks([]) # remove x axis label
ax.set_xlabel('x',fontsize=20) # set x label
ax.set_ylabel('pdf(x)',fontsize=20,rotation=0) # set y label
ax.xaxis.set_label_coords(0.61, -0.02) # x label coordinate
ax.yaxis.set_label_coords(-0.1, 0.5) # y label coordinate
plt.show()
```

insight: pada kurva diatas kita bisa melihat bahwa kemungkinan letak posisi di bidang x di angka 1

```python
fig, ax = plt.subplots()
x = np.arange(-4,4,0.001)
ax.set_title('N(0,$1^2$)')
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.plot(x,norm.pdf(x))
ax.set_ylim(0,0.45)
plt.show()
```

insight: pada data diatas kita bisa mengetahui bahwa Modality yang terbentuk adalah Unimodal, dan Skewnessnya Symmetric. Dan diketahui juga nilai rata-rata = 0, Standard Deviation = 1

## Normal Distribution PDF with Different Standard Deviations

```python
fig, ax = plt.subplots()
x=np.linspace(-10,10,100)
stdvs=[1.0,2.0,3.0,4.0]

for s in stdvs:
    ax.plot(x,norm.pdf(x,scale=s), label='stdv=%.1f' %s)

ax.set_xlabel('x')
ax.set_ylabel('pdf(x)')
ax.set_title('Normal Distribution')
ax.legend(loc='best', frameon=True)
ax.set_ylim(0,0.45)
ax.grid(True)
```

insight: semakin besar nilai standard deviation, maka kurva pada sumbu pdf(x) semakin mendekati 0 atau semakin kecil

## Using Random Normal Distribution Values

```python
fig, ax = plt.subplots()
xs = norm.rvs(scale=2,size=1000)
x = np.linspace(-10,10,100)
p = norm.pdf(x,scale=2)
v = np.var(xs)
m = np.mean(xs)

ax = fig.add_subplot(111)
ax.hist(xs, bins=10, alpha=0.5, density=True)
ax.plot(x,p, 'r-', lw=2)
ax.set_xlabel('x')
ax.set_ylabel('pdf(x)')
ax.set_title(f'mean={m:.2f}, var={v:.2f}')
ax.grid(True)
```

pada di step ini kita menggunakan data yang random jadi saat kita menjalankan step ini kemungkinan hasilnya bisa berbeda-beda. Tetpai, modality yang terlihat selalu berbentuk Unimodal

## Cumulative Normal Distribution Function

```python
fig, ax = plt.subplots()
#for distribution curve
x = np.arange(-4,4,0.001)
ax.plot(x, norm.pdf(x))
ax.set_title("Cumulative Normal Distribution")
ax.set_xlabel('x')
ax.set_ylabel('pdf(x)')
ax.grid(True)

#for fill_between
px = np.arange(-4,1,0.01)
ax.set_ylim(0,0.5)
ax.fill_between(px,norm.pdf(px),alpha=0.5, color='g')

#for text
ax.text(-1,0.1,"cdf(x)", fontsize=20)
plt.show()
```

pada step ini menunjukkan distribusi Cumulative dari var random X. Dan area yang diarsir menunjukkan probabilitas bahwa X kurang atau sama dari x

```python
from scipy.stats import norm
lessthan2=norm.cdf(x=2, loc=3, scale=2)
print(lessthan2)
```

```python
fig, ax = plt.subplots()

# for distribution curve
x= np.arange(-4,10,0.001)
ax.plot(x, norm.pdf(x,loc=3,scale=2))
ax.set_title("N(3,$2^2$)")
ax.set_xlabel('x')
ax.set_ylabel('pdf(x)')
ax.grid(True)

#for fill_between
px= np.arange(-4,2,0.01)
ax.set_ylim(0,0.25)
ax.fill_between(px,norm.pdf(px,loc=3,scale=2),alpha=0.5, color='g')

# for text
ax.text(-0.5,0.02,round(lessthan2,2), fontsize=20)
plt.show()
```

pada graph ini, diketahui bahwa rata ratanya adalah 3 dan Standard deviationnya 2.
