# Import Library

```python
import pandas as pd #analisa data
import numpy as np #operasi matematika, statistik, and matriks
import matplotlib.pyplot as plt #visualisasi data
import seaborn as sns #visualisasi data lebih menarik dari pyplot
sns.set()
```

# Data Preparation

```python
iris = pd.read_csv('iris.csv', names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class'])
iris.head()
```

# Visualisasi Data

## MatplotLib

### Scatter Plot

```python
fig, ax = plt.subplots()
ax.scatter(iris['sepal_length'], iris['sepal_width'])

ax.set_title('Iris Dataset')
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
```

insight: menurut graph ScatterPlot diatas, Most Sepal Width = 4.5, Most Sepal Length = almost 8.0

```python
colors={'Iris-setosa':'r', 'Iris-versicolor':'g', 'Iris-virginica':'b'}


fig, ax = plt.subplots()
for i in range(len(iris['sepal_length'])):
    ax.scatter(iris['sepal_length'][i], iris['sepal_width'][i],color=colors[iris['class'][i]])

ax.set_title('Iris Dataset')
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
```

perbedaanya dengan graph diatasnya, graph ini diberi warna untuk membedakan setiap class

### Line Chart

```python
columns = iris.columns.drop(['class'])
x_data = range(0, iris.shape[0])

fig, ax = plt.subplots()

for column in columns:
    ax.plot(x_data, iris[column])

ax.set_title('Iris Dataset')
ax.legend(["sepal_length","sepal_width","petal_length","petal_width"])
```

insight: Sepal lebih besar ukurannya, dri pada petal

```python
from numpy import sin
# consistent interval for x-axis
x = [x*0.1 for x in range(100)]

# function of x for y-axisb
y = sin(x)

# createline plot
plt.plot(x, y)
plt.show()
```

### Histogram

```python
g, ax = plt.subplots()
ax.hist(iris['sepal_width'])
ax.set_title('Wine Review Scores')
ax.set_xlabel('sepal_width')
ax.set_ylabel('Frequency')
```

insight: Lebar Sepal terbanyak antara 3.0-3.5

### Bar Chart

```python
fig, ax = plt.subplots()
data = iris['petal_length'].value_counts()

points = data.index
frequency = data.values

ax.bar(points, frequency)

ax.set_title('Iris Review Scores')
ax.set_xlabel('Petal_length')
ax.set_ylabel('Frequency')
```
